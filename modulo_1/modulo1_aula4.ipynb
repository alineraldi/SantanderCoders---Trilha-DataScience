```python
df = pd.read_csv('https://raw.githubusercontent.com/alineraldi/SantanderCoders---Trilha-DataScience/main/modulo_1/dataframes/california_housing.csv')
print(df.columns.tolist())
```

```python
# Renomear colunas
df = df.rename(columns={"date" : "Data", "price" : "Preço(R$)", "bedrooms" : "Quartos", "bathrooms": "Banheiros", "floors" : "Andares", "yr_built" : "Ano de construção", "zipcode" : "CEP"})
```

```python
df = df.drop(columns={"id", "sqft_living", "sqft_lot", "waterfront", "view", "grade", "condition", "sqft_above", "sqft_basement", "yr_renovated", "lat", "long", "sqft_living15", "sqft_lot15"})
df.head()
```

```python
# Tratar os números, remover espaços e pontos
def conversao_string(valor):
    return str(valor).replace(".", "").replace(" ", "").replace("T000000", "")

df["Data"] = df["Data"].apply(conversao_string)
df
```

```python
# Formatar a data

from datetime import datetime

def converter_data(valor):
    data_obj = datetime.strptime(str(valor), '%Y%m%d')
    data_formatada = data_obj.strftime('%d/%m/%Y')
    return data_formatada

df["Data"] = df["Data"].apply(converter_data)
df
```
