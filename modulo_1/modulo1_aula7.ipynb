```python
import pandas as pd
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/alineraldi/SantanderCoders---Trilha-DataScience/main/modulo_1/dataframes/california_housing.csv")
df
```

```python
df = df.drop(columns=["date"], axis=1)

# Preencher valores NaN com a média da coluna para todas as colunas numéricas
df.fillna(df.mean(), inplace=True)

df
```

```python
df.corr()
```

```python
x = df[["yr_built", "bedrooms", "sqft_living15"]]
y = df["price"]
```

```python
x_treino, x_teste, y_treino, y_teste = train_test_split(x, y)
```

```python
x_teste
```

```python
modelo = LinearRegression()
```

```python
modelo.fit(x_treino, y_treino)
```

```python
y_previsto = modelo.predict(x_teste)
```

```python
y_previsto
```

```python
df["price"].mean()
```

```python
mean_absolute_error(y_previsto, y_teste)
```
