```python
## Criando Gráficos

import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("https://raw.githubusercontent.com/alineraldi/SantanderCoders---Trilha-DataScience/main/modulo_1/dataframes/netflix_titles.csv")
df = df.drop(df.columns[0], axis=1)
```

```python
df.head()
```

```python
# Removendo NaN

dff = df.dropna()

dff
```

```python
eixo_x = dff["release_year"].value_counts().keys()
eixo_y = dff["release_year"].value_counts().values
```

```python
plt.bar(eixo_x, eixo_y)
plt.title("Filmes por ano")
plt.xlabel("Ano")
plt.ylabel("Quantidade de filmes")
plt.xticks(rotation=45)
plt.show()
```

```python
eixo_x = dff["rating"].value_counts().keys()
eixo_y = dff["rating"].value_counts().values
```

```python
plt.plot(eixo_x, eixo_y)

plt.title("Filmes por idade de restrição")
plt.xlabel("Restrição")
plt.ylabel("Quantidade de filmes")
plt.xticks(rotation=45)
plt.show()
```
