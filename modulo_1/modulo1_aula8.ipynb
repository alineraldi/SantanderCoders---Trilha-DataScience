```python
!pip install mlxtend

import pandas as pd

from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
from mlxtend.plotting import plot_confusion_matrix
from sklearn.metrics import confusion_matrix, classification_report
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/alineraldi/SantanderCoders---Trilha-DataScience/main/modulo_1/dataframes/breast_cancer_prediction.csv")
df.head()
```

```python
x = df.drop(columns=["id", "diagnosis"], axis=1)
```

```python
y = df["diagnosis"]
```

```python
x_treino, x_teste, y_treino, y_teste = train_test_split(x, y)
```

```python
modelo = LogisticRegression(max_iter=10000)
modelo.fit(x_treino, y_treino)
```

```python
y_previsto = modelo.predict(x_teste)
```

```python
cm = confusion_matrix(y_previsto, y_teste)
```

```python
plot_confusion_matrix(cm)
```

```python
print(classification_report(y_previsto, y_teste))
```
